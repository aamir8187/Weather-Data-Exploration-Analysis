```python
#Importing The libraries 

import pandas as pd
import numpy as np 
```

```python
#loading the data set into pandas dataframe 

df= pd.read_csv("weather_2012.csv")
df
```

```python
df.info()
```

# Here we can see Date/Time column is a objet column so we have to convert this into date time 

```python
df["Date/Time"]= pd.to_datetime(df["Date/Time"])
df.info()
```

```python
#let set the date/time column on index

df.set_index("Date/Time",inplace=True)
```

```python
# Find out the different weather conditions avialble 

df['Weather'].unique()
```

```python
df['Weather'].value_counts()
```

```python
#Find the number of times when the weather was excatly "clear".

len(df.loc[df['Weather']=='Clear'])
```

```python
#Find all the unique Wind Speed values recorded in the data set.

df["Wind Spd (km/h)"].unique()
```

```python
df['Wind Spd (km/h)'].value_counts()
```

```python
#Find the Number of times when wind speed is excatly 15km/hr.

len(df[df["Wind Spd (km/h)"]==15].value_counts())
```

```python
#Get the two columns from the data frame.

df[["Temp (C)","Weather"]].head(10)
```

```python
#first  10 to 20 rows of data frame 
df.iloc[10:20]
```

```python
#What were the first 5 pressure values recorded on Janaury 6?

df.loc["2012-01-06","Stn Press (kPa)"][:5]
```

```python
#Find all instances when Wind speed was above 30 and visibility was 25.

df[(df["Wind Spd (km/h)"]>30)&(df["Visibility (km)"]==25)]
```

```python
#Which were the top 10 hottest temp values and their counts?
df["Temp (C)"].value_counts().sort_values(ascending=False).head(10)
```

```python
#What is the mean temperature recorded by month?

mean_temp= df.groupby(df.index.month).agg(np.mean)
mean_temp["Temp (C)"]
```

```python
#Extract month name  from date and store it in new column.
df.reset_index(inplace=True)
df
```

```python
df['Month'] = df['Date/Time'].dt.strftime('%B')
df
```

# Get the list of months in which it rained with thunderstorm.



```python
#  Filter rows where it rained with thunderstorm
filtered_df = df[df['Weather'].str.contains('Rain') & df['Weather'].str.contains('Thunderstorm')]



# Get the unique list of months
months_with_rain_thunderstorm = filtered_df['Month'].unique().tolist()
```
